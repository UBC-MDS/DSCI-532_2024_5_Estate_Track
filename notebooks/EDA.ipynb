```python
df_num = df.groupby('City')['Number_Beds'].value_counts().reset_index()
df_num
# Assuming df_num is your DataFrame with 'City', 'Number_Beds', and 'count' columns
top_5_beds_per_city = df_num.groupby('City').apply(lambda x: x.nlargest(5, 'count')).reset_index(drop=True)

# top_5_beds_per_city
top_5_per_city = (df_num.sort_values(['City', 'count'], ascending=[True, False])
                                 .groupby('City')
                                 .head(5))
top_5_per_city
```
